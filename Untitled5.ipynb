**Word clouds**

```python
def make_wordcloud(text_series, title, filename, color):
    all_text = " ".join(text_series.dropna().astype(str))
    wc = WordCloud(width=800, height=400, background_color="white",
                   colormap=color, max_words=150).generate(all_text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")
    plt.title(title, fontsize=14)
    plt.tight_layout()
    plt.savefig(filename, dpi=150)
    plt.show()

make_wordcloud(fake_df["text"], "Most Common Words — Fake News",
               "/content/drive/MyDrive/NLP Assignment/wordcloud_fake.png", "Reds")

make_wordcloud(true_df["text"], "Most Common Words — Real News",
               "/content/drive/MyDrive/NLP Assignment/wordcloud_real.png", "Blues")
```

**Top 20 Most Frequent Words.**

```python
def get_top_words(text_series, n=20):
    all_words = " ".join(text_series.dropna().astype(str)).lower()
    all_words = re.sub(r"[^a-z\s]", "", all_words)
    words = all_words.split()
    stopwords = {"the","a","an","and","of","to","in","is","was","for",
                 "on","at","that","this","it","with","as","by","from","are"}
    words = [w for w in words if w not in stopwords and len(w) > 2]
    return Counter(words).most_common(n)

print("Top 20 words in FAKE news:")
fake_top = get_top_words(fake_df["text"])
for word, count in fake_top:
    print(f"  {word}: {count}")

print("\nTop 20 words in REAL news:")
real_top = get_top_words(true_df["text"])
for word, count in real_top:
    print(f"  {word}: {count}")
```

```python
# Install and download required NLP tools
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt_tab')

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
import string

print("All libraries loaded!")
```

 **lemmatizer and stopwords**

```python

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    # Step 1: Lowercasing
    text = text.lower()

    # Step 2: Remove special characters and punctuation
    text = re.sub(r'[^a-z\s]', '', text)

    # Step 3: Tokenization
    tokens = word_tokenize(text)

    # Step 4: Stopword removal
    tokens = [t for t in tokens if t not in stop_words]

    # Step 5: Lemmatization
    tokens = [lemmatizer.lemmatize(t) for t in tokens]

    return tokens

# Apply to a small sample of 5000 articles (faster for testing)
sample_df = df.sample(n=5000, random_state=42).copy()
sample_df['tokens'] = sample_df['text'].apply(preprocess_text)

# Preview results
print("ORIGINAL TEXT:")
print(df['text'].iloc[0][:200])
print("\nAFTER PREPROCESSING:")
print(sample_df['tokens'].iloc[0][:20])
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# Join tokens back into strings (sklearn needs strings not lists)
sample_df['clean_text'] = sample_df['tokens'].apply(lambda x: ' '.join(x))

# 1. Bag of Words (BoW)
bow_vectorizer = CountVectorizer(max_features=5000)
bow_matrix = bow_vectorizer.fit_transform(sample_df['clean_text'])

# 2. TF-IDF
tfidf_vectorizer = TfidfVectorizer(max_features=5000)
tfidf_matrix = tfidf_vectorizer.fit_transform(sample_df['clean_text'])

# 3. Bigram Model
bigram_vectorizer = CountVectorizer(ngram_range=(2,2), max_features=5000)
bigram_matrix = bigram_vectorizer.fit_transform(sample_df['clean_text'])

print("Feature Sizes:")
print(f"Bag of Words matrix shape:  {bow_matrix.shape}")
print(f"TF-IDF matrix shape:        {tfidf_matrix.shape}")
print(f"Bigram matrix shape:        {bigram_matrix.shape}")
```

**BoW, TF-IDF, Bigram**

```python
import numpy as np

# Top 10 words in BoW
bow_words = bow_vectorizer.get_feature_names_out()
bow_freq = np.asarray(bow_matrix.sum(axis=0)).flatten()
bow_top10 = sorted(zip(bow_words, bow_freq), key=lambda x: x[1], reverse=True)[:10]

# Top 10 words in TF-IDF
tfidf_words = tfidf_vectorizer.get_feature_names_out()
tfidf_scores = np.asarray(tfidf_matrix.sum(axis=0)).flatten()
tfidf_top10 = sorted(zip(tfidf_words, tfidf_scores), key=lambda x: x[1], reverse=True)[:10]

# Top 10 Bigrams
bigram_words = bigram_vectorizer.get_feature_names_out()
bigram_freq = np.asarray(bigram_matrix.sum(axis=0)).flatten()
bigram_top10 = sorted(zip(bigram_words, bigram_freq), key=lambda x: x[1], reverse=True)[:10]

print("Top 10 BoW Features:")
for word, freq in bow_top10:
    print(f"  {word}: {int(freq)}")

print("\nTop 10 TF-IDF Features:")
for word, score in tfidf_top10:
    print(f"  {word}: {round(score, 2)}")

print("\nTop 10 Bigram Features:")
for bigram, freq in bigram_top10:
    print(f"  {bigram}: {int(freq)}")
```
